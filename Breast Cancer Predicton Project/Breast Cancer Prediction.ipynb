# Breast Cancer Prediction Project

```python
# Importing Libraries
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

#### Data Collection & Analysis

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
# Checking for missing values
data.isnull().sum()
```

There are no NULL values in dataset

```python
# Statistical measures about the data
data.describe()
```

```python
data.info()
```

```python
data.drop( columns = ["Unnamed: 32"],inplace = True)
```

```python
# Checking the distribution of diagnosis label
data["diagnosis"].value_counts()
```

* B = Benign
* M = Malignant

```python
data.groupby("diagnosis").mean()
```

```python
label_mapping = { 'B':1, 'M':0}
data["diagnosis"] = data["diagnosis"].map(label_mapping)
```

```python
data.sample(5)
```

#### Separating Data into Features and Labels

```python
X = data.drop(columns = "diagnosis",axis =1)
y = data["diagnosis"]
```

```python
X
```

```python
y
```

#### Splitting the data into training data & test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

#### Standardize the Features

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

#### Model Training ( Logistic Regression )

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression model using Training data
model.fit(X_train,y_train)
```

#### Model Evaluation ( Accuracy Score) 

```python
# accuracy on training data
X_train_pred = model.predict(X_train)
accur = accuracy_score(y_train,X_train_pred)
```

```python
print("Accuracy on training data = ",accur)
```

```python
# accuracy on test data
X_test_pred = model.predict(X_test)
accur = accuracy_score(y_test,X_test_pred)
```

```python
print("Accuracy on test data",accur)
```

#### Building a Predictive System 

```python
input_data = np.array([893061, 11.6, 24.49, 74.23, 417.2, 0.07474, 0.05688, 0.01974,
       0.01313, 0.1935, 0.05878, 0.2512, 1.786, 1.961, 18.21, 0.006122,
       0.02337, 0.01596, 0.006998, 0.03194, 0.002211, 12.44, 31.62, 81.39,
       476.5, 0.09545, 0.1361, 0.07239, 0.04815, 0.3244, 0.06745]).reshape(1,-1)
scaled_input = scaler.fit_transform(input_data)
prediction = model.predict(scaled_input)
print(prediction)

if ( prediction[0] == 0):
    print("The Breast Cancer is Malignant")
else:
    print("The Breast cancer is Benign")
```

This project aims to predict breast cancer using a dataset with diagnostic features. After data analysis and preprocessing, a logistic regression model is trained to classify tumors as benign or malignant. Achieving high accuracy on training and test sets, the system can make accurate predictions for new tumors, distinguishing between benign and malignant cases.
